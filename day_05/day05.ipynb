```python
import requests
from bs4 import BeautifulSoup

def get_aoc_problem(day, year=2023):
    url = f"https://adventofcode.com/{year}/day/{day}"
    try:
        response = requests.get(url)
        response.raise_for_status()  # raises an exception for HTTP errors

        soup = BeautifulSoup(response.text, 'html.parser')
        
        problem_text = soup.find('article').get_text()
        return problem_text
    except Exception as e:
        return f"Error fetching problem: {e}"

day = 5
problem_prompt = get_aoc_problem(day)
print(problem_prompt)
```

```python
def calculate_location(seed_num):
    # return location num

    current = int(seed_num)

    for map_value_list in map_values:
        '''
        e.g. map_value_list = [[50, 98, 2], [52, 50, 48]]
        '''
        for line in map_value_list:
            '''
            e.g. line = [50, 98, 2] (list of integers)
            '''
            if line[1] <= current <= line[1] + line[2] - 1 :
                # within function window

                current += (line[0]-line[1])
                
                break
        # print(f'After line {line}, number = {current}')

    return current
```
