```python
with open('input.txt') as f:
    data = f.read()
    
lines = data.split('\n')[:-1]

games = {}

for line in lines:
    game_id = line.split(':')[0].split()[1]
    trials = line.split(':')[1].split(';')
    clean_trials = []
    for trial in trials:
        clean_trial = trial.strip().split(',')
        clean_trial = [string.strip() for string in clean_trial]
        clean_trials.append(clean_trial)
    games[game_id] = clean_trials

'''
12 red cubes
13 green cubes
14 blue cubes
'''

red_max = 12
green_max = 13
blue_max = 14

id_sum = 0

current_ball_count = {
    'red' : 0,
    'green' : 0,
    'blue' : 0
}

for game in games.keys():
    
    game_possible = True
    
    print(f"checking game: {game}")
    print(f"with trials {games[game]}")
    
    # check if game is possible
    
    game_trials = games[game]
    current_ball_count = reset_ball_counts(current_ball_count)
    print(f"Resetting to: {current_ball_count}")
    
    for trial in game_trials:
        print(f"trial: {trial}")
        for ball_count in trial:
            current_ball_count = update_ball_counts(ball_count,current_ball_count)
            print(f"Updating ball count to: {current_ball_count}")
            
        if not check_ball_counts(current_ball_count):
            print(f"Game not possible")
            print(f"{current_ball_count}")
            print(f"violates: (red_max,green_max,blue_max) = {(red_max,green_max,blue_max)}")
            game_possible = False
            break
        else:
            current_ball_count = reset_ball_counts(current_ball_count)
            print(f"Resetting to: {current_ball_count}")
            

    if game_possible:
        print('game is possible (adding to sum)')
        print(f'old sum = {id_sum}')
        id_sum += int(game)
        
    else:
        print('game is impossible')
        print(f'old sum = {id_sum}')
        
    print(f'end of game sum: {id_sum}')
    
print(id_sum)
```

```python
# Define a function to take input a string and dictionary.

# Update the dictionary and return 

def update_ball_counts(ball_count, current_ball_counts):
    
    count = int(ball_count.split()[0])
    colour = ball_count.split()[1]
    
    if colour == 'red':
        # red balls
        current_ball_counts['red'] += count
    
    if colour == 'green':
        # green balls
        current_ball_counts['green'] += count
    
    if colour == 'blue':
        # blue balls
        current_ball_counts['blue'] += count
    
    return current_ball_counts
```

```python
# Checks if ball count is legal

def check_ball_counts(current_ball_counts):
    
    
    if current_ball_counts['red'] > red_max:
        return False
    if current_ball_counts['green'] > green_max:
        return False
    if current_ball_counts['blue'] > blue_max:
        return False
    
    return True
```

```python
def reset_ball_counts(current_ball_counts):
    
    current_ball_counts['red'] = 0
    current_ball_counts['green'] = 0
    current_ball_counts['blue'] = 0
    
    return current_ball_counts
```

```python
# Define a function to take input a string and dictionary.

# Update the dictionary (according to rules of part 2) and return 

def update_ball_counts_2(ball_count, current_ball_counts):
    
    # print(f'\n Called with: ball_count = {ball_count} and current_ball_counts = {current_ball_counts}\n')
    
    count = int(ball_count.split()[0])
    colour = ball_count.split()[1]
    
    # print(f'count = {count}')
    # print(f'colour = {colour}')
    
    if colour == 'red':
        # red balls
        if current_ball_counts['red'] < count:
            current_ball_counts['red'] = count
    
    if colour == 'green':
        # green balls
        if current_ball_counts['green'] < count:
            current_ball_counts['green'] = count
    
    if colour == 'blue':
        # blue balls
        if current_ball_counts['blue'] < count:
            current_ball_counts['blue'] = count
    
    # print(f'returning: {current_ball_counts}')
    return current_ball_counts
```

```python
sum_of_powers = 0

current_ball_count = {
    'red' : 0,
    'green' : 0,
    'blue' : 0
}

current_best_count = {
    'red' : 0,
    'green' : 0,
    'blue' : 0
}

for game in games.keys():
    
    # print(f"checking game: {game}")
    # print(f"with trials {games[game]}")
    
    game_trials = games[game]
    
    current_ball_count = reset_ball_counts(current_ball_count)
    
    # print(f"Resetting to: {current_ball_count}")
    
    for trial in game_trials:
        # print(f"trial: {trial}")
        
        for ball_count in trial:
            # print(f'ball_count = {ball_count}')
            current_ball_count = update_ball_counts_2(ball_count, current_ball_count)
            # print(f"Updated ball count to: {current_ball_count}")
            
        # ['7 blue', '13 green']
        for colour in current_ball_count.keys():
            
            ball_count = current_ball_count[colour]
            input_str = str(ball_count) + ' ' + colour
            current_best_count = update_ball_counts_2(input_str, current_best_count)
            
        current_ball_count = reset_ball_counts(current_ball_count)    
        # print(f'current_best_count = {current_best_count} (at the end of trial = {trial})')
        
    # print(f'After all trials:')
    # print(f'Final cube total = {current_best_count}')
    
    cube_power = 1
    for colour in current_best_count.keys():
        cube_power *= current_best_count[colour]
    
    # print(f'sum_of_powers = {sum_of_powers}')    
    sum_of_powers += cube_power
        
    # print(f'end of sum_of_powers = {sum_of_powers}')
    
    current_best_count = reset_ball_counts(current_best_count)
    
print(sum_of_powers)
```
